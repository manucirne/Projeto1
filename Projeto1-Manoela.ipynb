## Novo parâmetro - Segurança alimentar

   Com as perguntas feitas aos moradores sobre o acesso aos alimentos foi possível chegar a uma única variável: Segurança Alimentar. Essa variável substituirá as respostas dadas anteriormente em uma nova coluna com duas possíveis respostas: 'Sim' e 'Não'.
   As perguntas que foram consideradas para criar o parâmetro de segurança alimentar foram as seguintes:
   
   -Os moradores tiveram a preocupação de que os alimentos acabassem antes de poderem comprar ou receber mais comida
   ]
   -Os alimentos acabaram antes que os moradores deste domicílio tivessem dinheiro para comprar mais comida
   
   -Os moradores ficaram sem dinheiro para ter uma alimentação saudável e variada
   
   -Os moradores deste domicílio comeram apenas alguns alimentos que ainda tinham porque o dinheiro acabou
    
    No momento em que as respostas foram separadas em intensidade, como por exemplo 'muita', 'pouca' ou 'regular' (segurança alimentar), percebeu-se que haviam apenas as respostas extremas e não as intermediárias. Ou seja, quem respondeu sim em alguma dessa perguntas respondeu sim para todas e vice-versa. Por esse motivo só existem duas opções em segurança alimentar: 'Sim' (tem segurança alimentar e 'Não' (não tem segurança alimentar).

```python
dados09 = naonulos09.rename(columns={"V4614":'RMDom','V2103':'PreAliAcab','V2105':'AliAcab','V2107':'SemAliSau','V2109':'SemDComeTem'})
dados09.loc[(dados09.AliAcab == 'Sim'),'Segali'] = 'Não'
dados09.loc[(dados09.AliAcab != 'Sim') & (dados09.PreAliAcab != 'Sim') & (dados09.SemDComeTem != 'Sim') & (dados09.SemAliSau != 'Sim'),'Segali'] = 'Sim'
```

## Segurança alimentar por região entre 2009 e 2013
  
  Os gráficos a cima mostram a diferença entre a porcentagem de respostas 'Sim' e 'Não' para segurança alimentar, em cada região do país entre 2009 e 2013. É possível observar que houve uma melhora nesse parâmetro, ou seja, a segurança alimentar é maior em 2013 que em 2009.
  
  No Centroeste essa mudança fica  mais evidente. Isso reflete que a falta de segurança alimentar nessa região é mais atrelada à renda que em outras, como o nordeste. A caatinga do nordeste impede que muitas pessoas tenham acesso a água e alimentos frescos devido ao clima árido característico. Essa questão ambiental não pode ser resolvida apenas com melhoras econômicas e distribuição de renda. A região norte tem uma questão diferente. O fato de ter mais difícil acesso impede que os alimentos cheguem ainda frescos e em quantidade suficiente à região. Por essa característica a questão da renda também não é o único fator que pode alterar o quadro da fome, apesar de ajudar na melhora. As regiões Sudeste e Sul, sendo as mais ricas e acessíveis do país tem mudanças significativas em relação à segurança alimentar quando a renda é aumentada e distribuida.
